```python
# Problem 1: Mean and std

# NOTE: I am assuming that <point> is int or float!
    
def findMean(point):
    num = len(point) #number of points
    sumPoints = 0    #initialize sume
    for p in point:
        sumPoints += p #sum all points
    return sumPoints/num #divide sum by number of points


def findStd(point):
    dim = len(point) #number of points
    mean = findMean(point) #call mean func
    diffSquared = 0 #initialize the difference squared
    for p in point:
        diffSquared += (p-mean)**2 #calculate difference squared between 
                                   #point and mean
    return (diffSquared/dim)**(1/2) #return std
    
```

```python
# Problem 2: Dot Product
def InnerProduct(list1, list2):
    dotProd = 0 #Initialize the dot product
    #If the lists are int/float and are the same length
    if  len(list1)==len(list2) and all((type(x)==int)|(type(x)==float) for x in list2) and all((type(x)==int)|(type(x)==float) for x in list2):
        for i in range(0, len(list1)):
            dotProd += (list1[i]*list2[i]) #Compute dotproduct
    else:
        print("Cannot compute dot product") #If not number and/or lists are not same length
        dotProd = 'NA'
            
    return dotProd
```

```python
# Problem 3: 
        
# Open and ready in file
with open('../Downloads/baby-names-data.txt') as f:
    lines = f.readlines()
lines = lines[1:] #Remove first row because it is just column names

#Generate lists of year, name, percent, and sex
years = []; name = []; perc = []; sex = []
for line in lines:
    row = line.strip("\n").split(",") #Split by commas
    years.append(row[0]) #Append to the relevant "list"
    name.append(row[1])
    perc.append(row[2])
    sex.append(row[3].replace('"', ''))  
years = list(years); name= list(name); perc = list(perc); sex = list(sex) #Set as lists


```

```python
#Part A:
# Most frequent boy names
eachYear = list(range(1880, 2009))
for year in eachYear:
    # Store name, percent, sex lists for correct year
    yearIndex = [i for i,y in enumerate(years) if y == str(year)] #Get index for each year
    namesYear  = [name[i] for i in yearIndex] #Create new lists with index for each year
    percYear   = [perc[i] for i in yearIndex]
    sexYear    = [sex[i]  for i in yearIndex]
    yearYear   = [years[i] for i in yearIndex]
    
    # Store name, percent, sex lists for boy
    boyIndex = [i for i, s in enumerate(sexYear)  if s =="boy"] # Get index for boy
    namesYear_boy  = [namesYear[i] for i in boyIndex] #Create new lists with index for boy names
    percYear_boy   = [percYear[i]  for i in boyIndex]
    sexYear_boy    = [sexYear[i]   for i in boyIndex]
    yearYear_boy   = [yearYear[i]  for i in boyIndex]
    
    # Print the boy name with the highest percent for each year
    print('Year: {} Name: {}'.format(year, namesYear_boy[percYear_boy == max(percYear_boy)]))
```

```python
#Part B:
# Most frequent girl names
eachYear = list(range(1880, 2009))
for year in eachYear:
    # Store name, percent, sex lists for correct year
    yearIndex = [i for i,y in enumerate(years) if y == str(year)] #Get index for each year
    namesYear  = [name[i] for i in yearIndex]#Create new lists with index for years
    percYear   = [perc[i] for i in yearIndex]
    sexYear    = [sex[i]  for i in yearIndex]
    yearYear   = [years[i] for i in yearIndex]
    
    # Store name, percent, sex lists for boy
    girlIndex = [i for i, s in enumerate(sexYear)  if s =="girl"] #Get index girls
    namesYear_girl  = [namesYear[i] for i in girlIndex]#Create new lists with index for girl names
    percYear_girl   = [percYear[i]  for i in girlIndex]
    sexYear_girl    = [sexYear[i]   for i in girlIndex]
    yearYear_girl   = [yearYear[i]  for i in girlIndex]
    
    # Print the boy name with the highest percent for each year
    print('Year: {} Name: {}'.format(year, namesYear_girl[percYear_girl == max(percYear_girl)]))
```

```python
# Part 3:
def countPop(nameList, percList, freq): 
    #Function creates a dictionary where the key is all names
    # and the values are the sum of percents 
    
    #For each name (item) and its index (i)
    #If the name is already in the dictionay, add the percent value
    #Otherwise create a new entry of that name, and let the value = percent 
    for i, item in enumerate(nameList): 
        if (item in freq): 
            freq[item] += float(percList[i])
        else: 
            freq[item] = float(percList[i])
    return(freq)

# For each year call countPop
freq = {} #Initialize dictionary
for year in eachYear:
    # Store name, percent, sex lists for correct year
    yearIndex = [i for i,y in enumerate(years) if y == str(year)] #Choose index with year
    namesYear  = [name[i] for i in yearIndex]
    percYear   = [perc[i] for i in yearIndex]
    sexYear    = [sex[i]  for i in yearIndex]
    yearYear   = [years[i] for i in yearIndex]
    # Call function to add to dictionary and update "count", where count is
    # actually the percent value
    freq = countPop(namesYear, percYear, freq)
# Find key with the max value
maximum = max(freq, key=freq.get)  
print('Most Popular Name:{} Sum Percent: {}'.format(maximum, freq[maximum]))
```
